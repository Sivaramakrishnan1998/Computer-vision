```python
#arithmetic operations are done to increase or decrease color intensity

import cv2
import numpy as np

img = cv2.imread('./images/input.jpg')

ones = np.ones(img.shape, dtype= 'uint8')*75 # img.shape used to construct a matrix 
# of same size multiplied by some scalar value

increased_intensity_img= cv2.add(img, ones)  

decreased_intensity_img= cv2.subtract(img, ones)

cv2.imshow("increased int",increased_intensity_img)
cv2.waitKey(0)

cv2.imshow("decreased int",decreased_intensity_img)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

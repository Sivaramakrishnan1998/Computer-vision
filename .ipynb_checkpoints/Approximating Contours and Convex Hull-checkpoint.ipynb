```python
import cv2
import numpy as np

img = cv2.imread('./images/house.jpg')
img2 = img.copy()
org_img = img.copy()
cv2.imshow('Original Image', org_img)
cv2.waitKey(0)

#convert img to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

#find contours
_, contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

#Iterate through the contours to find the bounding rectangle
for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(org_img, (x,y), (x+w,y+h), (0,0,255), 2)
    cv2.imshow('bounding rectangle on contours', org_img)
    cv2.waitKey(0)

# Approximating contours
# Iterate throught the contours to apply approxPolyDP() gives bounding points
for cnt in contours:
    accuracy = 0.01 * cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, accuracy, True)
    cv2.drawContours(img, approx, -1, (0,255,0), 3)
    cv2.imshow('Approximating Contour is done', img)
    
cv2.waitKey(0)

# using convex hull

# hull = [cv2.convexHull(cnt) for cnt in contours]
# cv2.polylines(img2, hull, True, (0, 255, 0), 2)
# cv2.imshow('Approximating contours joining lines', img2)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

    
```

```python
# Implementing Convex Hull
import cv2
import numpy as np

img = cv2.imread('./images/hand.jpg')
org_img = img.copy()

#display Original image
cv2.imshow('org img', img)
cv2.waitKey(0)

#convert to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh= cv2.threshold(gray, 176, 255, 0)

#find contours
_, contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

#sort contours by area
n = len(contours)-1
contours = sorted(contours, key = cv2.contourArea, reverse = False)[:n]

#Iterate through contours then find convex hull of it 
# for cnt in contours:
#     hull = cv2.convexHull(cnt)
hull = [cv2.convexHull(cnt) for cnt in contours]
cv2.drawContours(img, hull, -1, (0,255,0), 2) # second parameter must be an array object
cv2.imshow('Convex hull', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
